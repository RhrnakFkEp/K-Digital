```python
# Method Chaining
# . 을 이용해서 메소드를 연결해 여러개의 메소드를 한 번에 적용하는것
```

```python
df = pd.DataFrame({"A" : {0 : "a", 1 : "b", 2 : "c"}, 
                   "B" : {0 : 1, 1 : 3, 2 : 5}, 
                   "C" : {0 : 2, 1 : 4, 2 : 6}})
df
```

```python
# melt()

print(pd.melt(df, id_vars = ["A"], value_vars = ["B"]))
# id를 A로 잡고, 변수 기준은 B, 값은 변수의 값을 넣음
print()

df2 = pd.melt(df, id_vars = ["A"], value_vars = ["B", "C"])
print(df2)
print()

df3 = pd.melt(df, value_vars = ["A", "B", "C"])
print(df3)
```

```python
# method chaining을 사용하여 여러 함수 한 번에 사용

# melt, rename
df3 = pd.melt(df, value_vars=["A", "B", "C"]).rename(columns={
     "variable" : "var", "value" : "val"})
print(df3)
```

```python
df4 = pd.DataFrame({"bar" : ["A", "B", "C", "A", "B", "C"],
                    "baz" : [1, 2, 3, 4, 5, 6],
                    "foo" : ["one", "one", "one", "two", "two", "two"]})

df4
```

```python
# pivot()

print(df4.pivot(index = "foo", columns = "bar", values = "baz"))
# multi index
print()

# reset_index
df5 = df4.pivot(index = "foo", columns = "bar", values = "baz").reset_index()
print(df5)
```

```python
# melt()

df5.melt(id_vars = "foo", value_vars = ["A", "B", "C"]).sort_values(
    by = ["foo", "bar"]).rename(columns = {"value" : "baz"})
# sort_value by 생략가능
# 처음 형태로 변환됨
```

```python
s1 = pd.Series(["a", "b"])
s2 = pd.Series(["c", "d"])
print(s1)
print()
print(s2)
```

```python
# concat()
# concatenate

print(pd.concat([s1, s2],              
          keys = ["s1", "s2"],   
          names = ["Series name", "Row ID"]))
# join default = outer
# 시리즈로 합쳐짐. 데이터 프레임 X
print("\n\n")


df1 = pd.DataFrame([["a", 1], ["b", 2]],
                    columns = ["letter", "number"])
df2 = pd.DataFrame([["c", 3], ["d", 4]],
                    columns = ["letter", "number"])
print(df1)
print(df2)
print("\n\n")


print(pd.concat([df1, df2]))
```

```python
# concat() : inner join, outer join

df3 = pd.DataFrame([["c", 3, "cat"], ["d", 4, "dog"]],
                    columns = ["letter", "number", "animal"])
print(df3)
print("\n\n")


print(pd.concat([df1, df3]))
# outer join이라 NA 발생
# outer join은 합집합

print(pd.concat([df1, df3], join = "inner"))
# inner join은 교집합
```

```python
# concat() : sort

pd.concat([df1, df3],  sort = True)
# 정렬 기본값은 False임
```

```python
# concat() : verify_integrity
# 검증할것인지? 디폴트 False 
df5 = pd.DataFrame([1], index = ["a"])
df6 = pd.DataFrame([2], index = ["a"])
print(df5)
print(df6)

pd.concat([df5, df6], verify_integrity= True)
# 인덱스에 중복된 값이 있으면 에러가 발생하게끔 함
# 검증용 데이터로 많이 사용
# 마지막에 테스트 할 때 많이사용
```

```python
# merge()
# how : 
# left(left Join), right(Right Join), inner(Inner Join), outer(Outer Join)


# Filtering Join 
# 조건에 해당하는 값에 일치하는 것들만 걸러서 가져옴

# Indicator :
# default True. 어떻게 병합했는지 알려줌

# Query :
# 요청한 값만 가져옴
```

```python
adf = pd.DataFrame({"x1" : ["A", "B", "C"], "x2" : [1, 2, 3]})
print(adf)
print()

bdf = pd.DataFrame({"x1" : ["A", "B", "D"], "x3" : ["T", "F", "T"]})
print(bdf)
```

```python
# merge()

print("left join")
print(pd.merge(adf, bdf, how = "left", on = "x1"))
# how : default inner. 어떤 방식으로 병합할지 지정
# on : 지정한 키를 중심으로 병합
print()

print("right join")
print(pd.merge(adf, bdf, how = "right", on = "x1"))
print()

print("inner join")
print(pd.merge(adf, bdf, on = "x1"))
print()

print("outer join")
print(pd.merge(adf, bdf, how = "outer", on = "x1"))
print()
```

```python
# filtering joins

print(adf["x1"].isin(bdf["x1"]))
# adf.x1에 bdf.x1에 있는것이 들어있늕
# Ture, False 값 반환

# filtering
print(adf[adf["x1"].isin(bdf["x1"])])
# 일치하는 부분만 출력
print()

# ~ : not
print(adf[~adf.x1.isin(bdf.x1)])
# 일치하지 않는 부분만 출력
```

```python
# merge() indicator
# indicator default False

print(pd.merge(adf, bdf, how = "left", on = "x1", indicator = True))
# 새롭게 칼럼이 추가되고 해당 칼럼에 병합 방식이 기록됨
# both : 양쪽에 존재
# left only : 왼쪽에만 존재
# right only : 오른쪽에만 존재
```

```python
# query()
print(pd.merge(adf, bdf, how = "left", on = "x1", indicator = True).query(
    "_merge == 'left_only'"))
# 요청한 값만 가져옴


print(pd.merge(adf, bdf, how = "left", on = "x1", indicator = True
              ).query("_merge == 'left_only'"
                     ).drop(columns = ["_merge"]))
```

```python
df = sns.load_dataset("mpg")
df.head()
```

```python
# gorupby()

print(df.groupby(by = "origin"))
# origin 별로 나눔 값은 제대로 출력되지 않음
# 그룹으로 나누기만 하고 뭘 할건지 정해주지 않았기 때문
print("\n\n")

# size()
print(df.groupby(by = "origin").size())
# 그룹별로 사이즈(그룹별로 몇개가 존재하는지)를 계산
# 크기 순으로 나타남
# print(df["origin"].value_counts())
# 역순이기는 하지만 같은 결과
print("\n\n")


# min()
print(df.groupby(by = "origin").min())
# 그룹별 min 값 계산. 
# 전체 데이터를 그룹별로 나누고, 열별로 지정한 연산을 처리해서 출력함
print("\n\n")

# max()
# print(df.groupby(by = "origin").max())


# mean()
# print(df.groupby(by = "origin").mean())


# 그룹별로 계산하고 원하는 열부분만 출력
print(df.groupby(by = "origin")["cylinders"].median())

# 그룹 여러개 지정가능
print(df.groupby(["origin", "mpg"])["cylinders"].median())
```

```python
df2 = pd.DataFrame([[4, 7, 10], 
                    [5, 11, 8], 
                    [6, 9, 12]], 
                   index = [1, 2, 3], columns = ["a", "b", "c"])
print(df2)
```

```python
# shift()

# 아래로 shift
print(df2.shift(1))
# 아래로 한 행씩 밀림
# 생기는 공백은 Na 값으로 채워짐
print()

# 위로 shift
print(df2["a"].shift(-1))
# 위로 한 행씩 밀림
# 공백은 Na
```

```python
# rank

print(df["model_year"].rank(method = "min"))
# method : 어떤방식으로 랭크를 매길건지
# 오름차순으로 
print("\n\n")

print(df["model_year"].rank(pct = True))
# pct : 각 값이 차지하는 비율 출력
print("\n\n")


print(df["model_year"].rank(method = "first").head())
```

```python
# cum
# cumsum, cummax, cummin, cumprod

print(df2)
print("\n\n")

# cumsum() : 누적합
print("cumsum()")
print(df2.cumsum())
# cumsum 1행 = 1행(df2), 
# cumsum 2행 = 1행(df2) + 2행(df2), 
# cumsum 3행 = 1행(df2) + 2행(df2) + 3행(df2) ....
print("\n\n")

# cummax() : 최대
print("cummax()")
print(df2.cummax())
# cummax 1행 = 1행(df2)
# cummax 2행 = cummax 1행 vs 2행(df)
# cummax 3행 = cummax 2행 vs 3행(df) ....
print("\n\n")

# cummin() : 최소
print("cummin()")
print(df2.cummin())
print("\n\n")

# cumprod() : 누적곱
print("cumprod()")
print(df2.cumprod())
print("\n\n")
```
