### seaborn

```python
iris.head()
```

```python
# pair plot
sns.pairplot(data = iris)
# 같은 데이터끼리는 히스토그램
# 다른 데이터끼리는 scatter plot
```

```python
# pair plot
sns.pairplot(data = iris, hue = "species",             # hue로 색 구분
             vars = ["petal_length", "petal_width"],   # vars로 원하는 데이터만 추출해서 보여줌
             height = 4)  # 플랏 높이(크기) 키워줌
```

```python
# joint plot
sns.jointplot(x = "petal_length", y = "petal_width", data = iris, 
              kind = "hex")  
# kind에 따라 그래프가 다르게 그려짐
# scatter : 산점도
# reg : 회귀선
# kde : 확률밀도. 등고선
# hex : 육각타일. 확률밀도와 유사하지만 더 뭉뚱그려보여주는듯?
```

```python
# lm
sns.lmplot(x = "sepal_length", y = "sepal_width", hue = "species", data = iris)
# lm: linear model? 인듯? 스케터 위에 회귀선 그려줌
# hue를 기준으로 구분해서 그려줌
```

```python
# heatmap
sns.heatmap(iris.corr(), annot = True, cmap = "Reds")
# 상관계수에 따라 히트맵 그림
# annot : 수치정보 표시. Flase 가 디폴트
# cmap : 색상 변경. Reds, Blues, Greens...
```

```python
# scatter plot
sns.scatterplot(x = "petal_length", y = "petal_width", size = "species", data = iris)
# size : 점 크기
```

```python
# relplot. 산점도
# 종별로 색 다르게 해서 나타낼 수 있음
sns.relplot(x = "petal_length", y = "petal_width", hue = "species", data = iris)
```

```python
tips.head()
sns.pairplot(tips, hue = "size", height = 3)
```

```python
# barplot
sns.barplot(x = "day", y = "tip", data = tips, hue = "smoker")
# hue로 하나의 바 플랏을 분할할 수 있음
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "size", col_wrap = 2, data = tips)
# col을 기준으로 플랏을 여러개로 분할함
# col_wrap으로 한 줄에 몇개씩 볼건지 지정
```

```python
# pivot table
# 통계정보를 그룹바이랑 유사하게 만들어줌
pivot = tips.pivot_table(index = "day", columns = "smoker", values = "tip")
print(pivot)
```

```python
sns.heatmap(pivot, annot = True, cmap = "Blues")
```

```python
sns.scatterplot(x = "total_bill", y = "tip", size  = "size", data = tips)
```

```python
sns.relplot(x = "total_bill", y = "tip", hue = "day", 
            col = "smoker", col_wrap = 2,
            data = tips)
```

```python
# catplot
# 카테고리별로 구분해서 그려줌
sns.catplot(x = "day", y = "total_bill", data = tips)
```
