```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup

#----------------------------------------------------------------
driver = webdriver.Chrome("./chromedriver")

# 페이지 열릴 때마다 딜레이 시켜줌
# 명시적으로 3초를 기다림
# 인터넷 속도(네트워크 상태)가 빠르면, 딜레이가 줄어듦
# 무언가 정보가 들어오면 wait을 그만둔단것.
driver.implicitly_wait(3)  # time 과연결.
driver.get("https://www.naver.com")

# URL에 연결
html = driver.page_source
# 파서
html_parsing = BeautifulSoup(html, "lxml")

# driver.close()
```

```python
# 네이버 로그인 버튼 클릭
driver.find_element_by_xpath('//*[@id="account"]/a').click()
# 넘어간 페이지
# https://nid.naver.com/nidlogin.login?mode=form&url=https%3A%2F%2Fwww.naver.com


# 네이버 아이디와 비밀번호
# 본인 id와 pw 넣어주면됨.
naver_id = "2a4310"
naver_password = "DLftkddyd_0101"


# 네트워크 상황에 따라 필요. 고정된 시간을 기다려줌.
# 고정!된시간임.
time.sleep(3)



# 클래스는 겹치는게 있어서 하면안됨.
# id 영역
# 자신의 네이버 id 값 입력.
# 해당 웹페이지를 변경시킬 수 있는 드라이버를 꼭 써줘야함.
input_id = driver.find_element_by_id("id")
# 어떤 값이 들어가 있을 수 있기 때문에 clear 해줘야햠.
# 값이 안들어가있으면 안넣어도 됨. clear()
input_id.clear()
input_id.send_keys(naver_id)  # naver_id를 input_id에 넣음


# password 영역
# 자신의 네이버 PW 값 입력
# html id 속성
input_pw = driver.find_element_by_id("pw")
input_pw.clear()
input_pw.send_keys(naver_password)


# 로그인 버튼 클릭
driver.find_element_by_xpath('//*[@id="log.login"]').click()



# 셀레늄봇을 통해 크롤링으로 아이디 비밀번호 자동으로 입력해서
# 들어가려고 시도했고, 이것은 서버에 자원을 계속 요청하므로
# 서버 자원 낭비 등을 방지하기 위해서 캡챠, 보안 등에 해당하는 것을 만듦.
# 자바스크립트를 통해 뚫을 수 있음.


# 결론 : 셀레늄으로 안들어가짐. 보안상의 문제.
```

```python
# 네이버 로그인 버튼 클릭
driver.find_element_by_xpath('//*[@id="account"]/a').click()
# 넘어간 페이지
# https://nid.naver.com/nidlogin.login?mode=form&url=https%3A%2F%2Fwww.naver.com


# 네이버 아이디와 비밀번호
# 본인 id와 pw 넣어주면됨.
naver_id = "2a4310"
naver_password = "DLftkddyd_0101"


# 네트워크 상황에 따라 필요. 고정된 시간을 기다려줌.
# 고정!된시간임.
time.sleep(3)



# 자바스크립트로 보안을 뚫고 들어가보자

# 자바 명령어를 넣어주는것.
# 자바스크립트 자체의 명령어를 넣는것임
# 자바처럼 동적인것에 대한것은 아직 보안상 막을 수 없음
# 'id ': html에서의 id 속성값임.
driver.execute_script("document.getElementsByName('id')[0].value=\'" + naver_id + "\'")
driver.execute_script("document.getElementsByName('pw')[0].value=\'" + naver_password + "\'")



# 로그인 버튼 클릭
driver.find_element_by_xpath('//*[@id="log.login"]').click()

# 오 개신기 이게 들어가네
```
