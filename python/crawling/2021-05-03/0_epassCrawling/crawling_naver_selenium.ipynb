```python
# 기사 내용 가져오기
# 기사 내용 전체
# print(html_parsing)



# title : 기사의 제목
title = html_parsing.find("h2", {"class" : "title"})
# print(title)
# print(title.text)
print(f"기사 제목 : {title.text}\n\n")



# 기사 내용(텍스트) :
# find_all() : 찾아서 가져올 것이 여러개(2개이상)일때 사용
# contents = html_parsing.find_all("p", {"class" : "editor-p read"})
# 안가져와짐. find_all 에서는 클래스 하나만 가져올 수 있음
contents = html_parsing.find_all("p", {"class" : "editor-p"})
# print(contents)

# print(contents.text)
# 여러개의 엘리먼트(원소)를 가져서 텍스트를 추출할 수 없음. for 구문 사용해야함.

print("기사 내용 :")
for content in contents:
    try :
        # content 있을 때는 시도
        print(content.text + "\n")
    except :
        # content 없을 때는 에러 출력.
        print("error")

        
        
# 기자이름과 이메일
reporter = html_parsing.find("span", {"class":"name"})
# print(reporter)

reporter_name = reporter.find("strong")
# print(reporter_name)
# print(reporter_name.text)
reporter_name = reporter_name.text
reporter_name = reporter_name.strip()  # 앞뒤 공백 없애는 함수
print(f"기자 이름 :{reporter_name}")

reporter_email = reporter.find("span", {"class" : "email"})
# print(reporter_email)
# print(reporter_email.text)
print(f"기자 메일: {reporter_email.text.strip()}")
# 위처럼 나누지 않고, 한방에도 가능.

    

# 기사 내용(이미지)
# img src = https://newsimg.hankookilbo.com/cms/articlerelease/2021/04/29/18c51f44-4c96-44aa-af0c-0271cc001acc.jpg"
# tag : div     class : editor-img-box read
image_all = html_parsing.find_all("div", {"class" : "editor-img-box"})
# 이미지를 가져오지 않고, 이미지의 URL을 가져옴.
# print(image_all)

idx = 1

for image in image_all:
    image_result = image.find("img", {"class":"read"})
#     print(image_result) # 많은 태그는 없어지고 이미지의 URL 하나만 남음.
    image_result = image.find("img")["src"]
    print(str(idx) + "번째 이미지 : ", image_result)
    idx += 1


```
