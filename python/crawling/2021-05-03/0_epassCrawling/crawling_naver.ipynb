```python
'''
작성일자 : 2021-0429



'''
# 기본 작성부분.
# 프레임워크에 해당?
import requests
import bs4

# requests.get("URL")
req = requests.get("https://www.naver.com")
# http : hyper text transper protocol
print(req)   # http code 출력. Response [200] : 정상적으로 응답이 됨을 의미.
# request : 요청
# response : 응답
# 사이트에 연결함을 서버에게 요청
# 서버는 클라이언트에게 해당 요청에 대해 응답
# print(req.text)   # 네이버 html 코딩 코드에 해당하는 것 역시 가져옴.


naver_html = req.text

bs_obj = bs4.BeautifulSoup(naver_html, "lxml")
print(bs_obj)
# 1. 메일~
```

```python
# 1. 메일~TV  list
# F12 개발자도구

# 무언가를 찾을것임.
# class "list_nav type_fix"
# 클래스가 2개 지정되어있음.
# {} : 브레이스괄호
ul = bs_obj.find("ul", {"class" : "list_nav"})   # find는 가장 첫번째 태그만 가져옴
# clsss : 클래스명
# print(ul)

li_list = ul.findAll("li")  # 리스트 형으로 가져옴
# print(li_list)
# print(li_list[0])

# print(li_list.text)
# 안뽑아짐. 리스트형태라 각각 하나씩 가져와야함
# print(li_list[0].text)

# for n in li_list:
#     print(n.text)
for li in li_list:
    a_tag = li.find("a")
    print(a_tag.text)
```

```python
# 2. 사전~웹툰  list

ul = bs_obj.find("ul", {"class" : "NM_FAVORITE_LIST"})
# print(ul)

li = ul.findAll("li")
# print(li)

# for n in li:
#     aTag = n.find("a")
#     print(a_tag.text)

# print(li[0])

for n in li:
    aTag = n.find("a")
    print(aTag.text)
```
